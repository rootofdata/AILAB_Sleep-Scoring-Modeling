```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
import json 
def open_json_file(filename): 
    with open(filename, encoding='UTF8') as file: 
        try: # 파싱 성공시 데이터 리턴 
            return json.load(file) 
        except ValueError as e: 
            print('파싱 실패! 에러 코드: {}'.format(e)) # 파싱 성공시 None 리턴 
            return None
```

```python
def answer(filename):
    if __name__ == "__main__": 
        json_data = open_json_file(filename)
    max_dat = list() ; min_dat = list()
    for i in range(len(json_data)):
        max_dat.append(json_data[i]['spo2_max'])
        min_dat.append(json_data[i]['spo2_min'])
    max_dat = np.array(max_dat) ; min_dat = np.array(min_dat)
    avg_dat = (max_dat + min_dat) / 2
    plt.figure(figsize = (20,10))
    plt.plot(range(len(json_data)), max_dat , color = 'red')
    plt.plot(range(len(json_data)), min_dat , color = 'green')

    
```
